Ethnic Group Counts

```python
plt.figure(figsize=(10,10))

ax = sns.countplot(y = 'EthnicGroup', orient = 'h', data = data)

for i in ax.containers:
    ax.bar_label(i,)
    
plt.show()
```

Gender Distribution

```python
cx = sns.countplot(x = 'Gender', data = data, palette = 'Set2')

for i in cx.containers:
    cx.bar_label(i,)

plt.show()
```

Aggregation data 

```python
aggregated = data.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(aggregated)
```

```python
plt.figure(figsize=(7,7))

sns.heatmap(aggregated.corr())

plt.show()
```

Check outliers in Math score, reading score and writing score

```python
plt.figure(1)

sns.boxplot(data = data1, x = "MathScore")
plt.title('Math Score Outlier')

plt.figure(2)
sns.boxplot(data = data1, x = "ReadingScore")
plt.title('Reading Score Outlier')

plt.figure(3)
sns.boxplot(data = data1, x = "WritingScore")
plt.title('Writing Score Outlier')

plt.show()
```

The school wants give extra marks to all the students who study more than 10 hours

```python
plt.figure(figsize = (6,6))
plt.subplot(1,1,1)
tx = sns.histplot(data['WklyStudyHours'] , bins = 20)
for i in tx.containers:
    tx.bar_label(i,)

plt.show()
```
